```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Observer:
    
    def __init__(self):
        pass
    
    def plot_chart(self, df):
        ax = plt.gca()
        
        df.plot(kind='line',x='index',y='min',color='green', ax=ax)
        df.plot(kind='line',x='index',y='average', color='red', ax=ax)
        df.plot(kind='line',x='index',y='max', color='blue', ax=ax)

        plt.show()
        
```

```python
numpy_data = np.array([[1, 1, 2, 3], [2, 4, 5, 6], [3, 7, 8, 9]])

df = pd.DataFrame(data=numpy_data, index=[1,2,3], columns=["index", "min", "average", "max"]) 
print(df)
observer = Observer()

observer.plot_chart(df)
```
